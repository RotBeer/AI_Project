### 피싱 사이트란?
 유명 사이트를 위장한 가짜 사이트를 통해 사용자의 로그인 ID, 패스워드, 신분증 촬영 등 개인정보를 탈취해 각종 공격을 하는 사이트입니다.
 
 주로 금전적 피해를 일으키는 공격을 하는데 예금인출, 결재 대납, 명의 도용 등의 피해를 줍니다.

 동시에 해커는 탈취한 정보로 진짜 사이트에 접속하여 가상화폐 등의 자산을 탈취합니다.

### 피싱사이트 접속 예방 수칙?
1. 인터넷 주소창에 자물쇠가 있는지 꼭 확인합니다. 
2. 인터넷 주소(URL)가 정확한지 확인합니다. 
3. 이메일 발신자 주소를 확인합니다. 
4. 이메일에서 버튼 링크로 넘어올 때 특히 더 주의합니다. 
5. 인터넷 검색보다는 직접 주소를 입력하거나 즐겨찾기를 이용합니다.

### 하지만...
피싱사이트에 의한 피해는 **무의식**중에 **의심없이** 인터넷을 사용하다가 발생합니다.

모든 사람이 '피싱사이트 접속 예방 수칙'을 외우고 다니면서 웹사이트에 접속할 때 마다 의심하며 사용할 수 는 없습니다.

무작위 대상을 공격하거나, 특정 개인이나 회사를 대상으로 하거나, 고위 경영 간부나 비즈니스 내 기타 고위직을 대상으로 하는 등 종류도 다양해졌습니다.

웹사이트를 만드는 기술 이외에 특별한 기술이 필요하지 않으며, 유명사이트를 똑같이 복제하는 것도 어렵지 않으며, 백신에도 검출 되지 않습니다.

이러한 특징들 때문에 피싱사이트들은 개인 및 금융 정보를 탈취하는 가장 성공적이고 효과적인 방법이 되었습니다.

### 그렇다면!
머신러닝을 가짜 사이트를 분류해내거나, 가짜 사이트일 확률이 높다면 사용자에게 경고하는 방식으로 사용 할 수 있을 것으로 보입니다.

# Part 2. 모델링 및 모델 해석
## 모델링
주어진 데이터로 모델링을 해보겠습니다.

```python
from sklearn.model_selection import train_test_split

features = ['length_url', 'length_hostname', 'nb_dots', 'nb_hyphens', 'nb_at', 'nb_qm', 'nb_and', 'nb_eq', 'nb_underscore',
 'nb_tilde', 'nb_percent', 'nb_slash', 'nb_colon', 'nb_semicolumn', 'nb_space', 'nb_www',
 'nb_com', 'nb_dslash', 'http_in_path', 'ratio_digits_url', 'ratio_digits_host', 'nb_redirection', 'shortest_word_host', 'shortest_word_path',
 'longest_word_host', 'longest_word_path', 'avg_word_host', 'avg_word_path']
target = 'status'

X = df_cp[features]
y = df_cp[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.metrics import classification_report

baseline = len(y_train) * [y_train.mode()[0]]
print(classification_report(y_train, baseline))
```

```python
from xgboost import XGBClassifier
from scipy.stats.distributions import uniform
from sklearn.model_selection import RandomizedSearchCV

xgb = XGBClassifier(
        objective="binary:logistic",
        eval_metric="error",
        n_estimators=200,
        random_state=42,
        n_jobs=-1,
        learning_rate=0.1,
        use_label_encoder=False,
    )

params = {
    "xgbclassifier__max_depth": [2, 4, 6, 8],
    "xgbclassifier__min_child_weight": [2, 4, 8, 16],
    "xgbclassifier__colsample_bytree": uniform(
        loc=0.5, scale=0.5
    ),  # 0.5 ~ 1 사이의 uniform 분포로 범위를 지정해 줍니다.    
}

randomized_search = RandomizedSearchCV(
    xgb,
    param_distributions=params,
    scoring="roc_auc",
    n_iter=10,
    cv=5,
    verbose=3,
    random_state=42,
)
randomized_search.fit(X_train, y_train)
print("최적 하이퍼파라미터: ", randomized_search.best_params_)
print("최적 AUC: ", randomized_search.best_score_)
```

```python
print(classification_report(y_train, randomized_search.predict(X_train)))
```

```python
import eli5
from eli5.sklearn import PermutationImportance

# permuter 정의
permuter = PermutationImportance(
    randomized_search,  # model
    scoring="accuracy",  # metric
    n_iter=10,  # 다른 random seed를 사용하여 100번 반복
    random_state=42,
)

# 스코어를 다시 계산합니다.
permuter.fit(X_train, y_train)

feature_names = X_test.columns.tolist()
pi = pd.Series(permuter.feature_importances_, feature_names).sort_values()
pi.plot.barh()
plt.title("Permutation Importance")
```

```python
from pdpbox.pdp import pdp_isolate, pdp_plot

target_feature = "nb_www"

isolated = pdp_isolate(
    model=randomized_search,
    dataset=X_train,
    model_features=X_train.columns,
    feature=target_feature,
    grid_type="percentile",  # default = 'percentile', or 'equal'
    num_grid_points=10,  # default = 10
)
pdp_plot(isolated, feature_name=target_feature)
```
